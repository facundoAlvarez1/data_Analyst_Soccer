```python
from statsbombpy import sb
import pandas as pd
import matplotlib.pyplot as plt
from mplsoccer import VerticalPitch, Pitch
```

```python
sb.competitions()
```

```python
sb.matches(competition_id=43,season_id=106).sort_values(by='match_date')
```

```python
partido = sb.events(match_id=3869685)
```

```python
partido.columns
```

```python
partido.player.value_counts()
```

```python
jugador = partido[partido['player']== 'Damián Emiliano Martínez']
```

```python
jugador[['x','y']] = jugador['location'].apply(pd.Series)
```

```python
fig, ax = plt.subplots(figsize=(16,9))
pitch = Pitch(
    pitch_type='statsbomb',
    axis=True, 
    label=True, 
    tick=True,
    goal_type='box'
)

pitch.draw(ax=ax)
```

```python
jugador.type.value_counts()
```

```python
pases= jugador[jugador['type'] == 'Pass']
pases[['x','y']] = pases.location.apply(pd.Series)
pases[['endX','endY']] = pases.pass_end_location.apply(pd.Series)

pases[['x','y','endX','endY']] 
```

```python
fig, ax = plt.subplots(figsize=(16,9),facecolor='#c0c0c0')
pitch = Pitch(
    pitch_type='statsbomb',
    axis=True, 
    label=True, 
    tick=True,
    goal_type='box',
    pitch_color='grass',
    line_color='white',
    stripe=True
)

pitch.draw(ax=ax)
pases_completos = pases[~pases['pass_outcome'].isin(['Incomplete','Pass Offside'])]
pases_incompletos = pases[pases['pass_outcome'].isin(['Incomplete','Pass Offside'])]

pitch.arrows(pases_completos['x'],pases_completos['y'],pases_completos['endX'],pases_completos['endY'],ax=ax ,label = 'Pases Completos',color ='#2980B9')
pitch.arrows(pases_incompletos['x'],pases_incompletos['y'],pases_incompletos['endX'],pases_incompletos['endY'],ax=ax ,label = 'Pases Incompletos',color = '#EC7063')

total_pases = len(pases)
porcentaje_pases_completos = len(pases_completos) / total_pases * 100
porcentaje_pases_incompletos = len(pases_incompletos) / total_pases * 100

leyenda_texto = (
    f'Pases Completos: {porcentaje_pases_completos:.2f}%\n'
    f'Pases Incompletos: {porcentaje_pases_incompletos:.2f}%'
)
ax.text(1.05, 0.5, leyenda_texto, transform=ax.transAxes, fontsize=12, verticalalignment='center')

# Añadir título
ax.set_title("Análisis de Pases Emiliano Martinez en final mundial qatar 2022", fontsize=18, fontweight='bold', color='#333333')

# Añadir etiquetas de ejes con estilo
ax.set_xlabel("Coordenada X", fontsize=14, color='#333333')
ax.set_ylabel("Coordenada Y", fontsize=14, color='#333333')

# Mostrar la leyenda
ax.legend(fontsize=12)

# Mostrar el gráfico
plt.show()
```
