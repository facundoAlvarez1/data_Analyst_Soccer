```python
df_plot = df[['Equipo', 'T/90', 'G/T']]
df_plot.Equipo = df_plot.Equipo.replace({
   'Tucumán' : 'Atletico Tucumán' ,
   'CA Huracán' : 'Huracan' , 
   'CA Unión' : 'Union' , 
   'Club Atlético Belgrano' : 'Belgrano',
   'Instituto Atlético Central Córdoba' : 'Instituto'
})
df_plot = df_plot.sort_values(by = 'Equipo')
df_plot
```

```python
x = df_plot['T/90']
y = df_plot['G/T']
nombreEjeX = 'Tiros Totales por 90 min' 
nombreEjeY = 'Goles Por Tiro' 

promx = x.mean() #calcular la media de eje X 
promy= y.mean() #calcular la media de eje Y
color_plot = 'white'
facecolor = '#292323'

fig , ax = plt.subplots(figsize =(16,9))

ax.scatter(x,y)
ax.axhline(promy,color = color_plot)
ax.axvline(promx , color = color_plot)


#Detalles esteticos
#Setea distintos las lineas de borde
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_color(color_plot)
ax.spines['left'].set_color(color_plot)

#Setear Colores de Fondo
fig.patch.set_facecolor(facecolor)
ax.set_facecolor(facecolor)

#Cambiar las lineas de bordes
for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_linewidth(1.5)
ax.yaxis.set_tick_params(pad = 12,colors=color_plot,labelsize=12)
ax.xaxis.set_tick_params(pad = 12,colors=color_plot,labelsize=12)

ax.set_xlabel(nombreEjeX , color = color_plot , size = 16)
ax.set_ylabel(nombreEjeY , color = color_plot , size = 16)

ax.set_title(f'{nombreEjeX} vs. {nombreEjeY} - Torneo Binance 2023', fontproperties=titulo.prop, color=color_plot, fontsize=16, pad=20)

def getImage(path, zoom=0.24):
    img = Image.open(path)
    img.thumbnail((int(img.width * zoom), int(img.height * zoom)))
    return OffsetImage(img, zoom=zoom)
import os

files = os.listdir('C:/Users/USUARIO/OneDrive/Escritorio/data_Analyst_Soccer/data_Analyst_Soccer/logos1eraDivisionArgentina')

for x0,y0,file in zip(x,y,files):
    ab = AnnotationBbox(getImage(f'C:/Users/USUARIO/OneDrive/Escritorio/data_Analyst_Soccer/data_Analyst_Soccer/logos1eraDivisionArgentina/{file}',zoom= 0.24),(x0,y0), frameon= False)
    ax.add_artist(ab)

    plt.savefig('disparosXequipo.png', dpi = 300 , bbox_inches ='tight')
```
